```python
import nltk
nltk.download("book")
from nltk.book import *
from nltk.corpus import brown
```

```python
import math

# Cache for IDF values to improve performance
idf_cache = {}

def tf(term, text):
    """Calculate the term frequency (TF) of the term in the given text."""
    return text.count(term) / len(text) if len(text) > 0 else 0

def idf(term, texts):
    """Calculate the inverse document frequency (IDF) of the term across the given corpus of texts.

    The number of texts in the corpus divided by the number of texts that the term appears in.
    If a term does not appear in any of the texts, returns 0.0.
    """
    # Use cached IDF value if available
    if term in idf_cache:
        return idf_cache[term]

    # Count the number of documents containing the term
    matches = len([True for text in texts if term in text])

    # Calculate IDF
    if len(texts) == 0:
        raise ValueError("IDF undefined for an empty document collection")
    idf_value = math.log(len(texts) / matches) if matches > 0 else 0.0

    # Cache the result
    idf_cache[term] = idf_value
    return idf_value

def tf_idf(term, text, texts):
    """Calculate the TF-IDF value for a term in a given text, with respect to a corpus of texts."""
    return tf(term, text) * idf(term, texts)

# Example usage with the Brown corpus
from nltk.corpus import brown

# Convert the Brown corpus into a list of texts, where each text is a list of words
corpus = [brown.words(fileid) for fileid in brown.fileids()]

# Test IDF for the term "today" in the Brown corpus
idf_value = idf("thoroughly", corpus)
print(f"IDF for 'thoroughly' in the Brown corpus: {idf_value}")

# Choose a sample text from the Brown corpus (e.g., the first document)
sample_text = brown.words(brown.fileids()[0])

# Calculate TF for the term "today" in the sample text
tf_value = tf("thoroughly", sample_text)
print(f"TF for 'thoroughly' in the first document of the Brown corpus: {tf_value}")

# Calculate TF-IDF for the term "today" in the sample text
tf_idf_value = tf_idf("thoroughly", sample_text, corpus)
print(f"TF-IDF for 'thoroughly' in the first document of the Brown corpus: {tf_idf_value}")
```
